```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv') 

# Agregar estas líneas para investigar:
print("Valores de odometer:")
print(car_data['odometer'].describe())
print("\nValores de price:")
print(car_data['price'].describe())
print("\nCantidad de valores nulos:")
print("Odometer nulos:", car_data['odometer'].isnull().sum())
print("Price nulos:", car_data['price'].isnull().sum())
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

car_data = pd.read_csv('../vehicles_us.csv')
fig = px.histogram(car_data, x="odometer")
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Configurar para VS Code
pio.renderers.default = "browser"

car_data = pd.read_csv('../vehicles_us.csv')
fig = px.histogram(car_data, x="odometer")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv')  # Nota la ruta relativa
fig = px.scatter(car_data, x="odometer", y="price")
fig.show()
```
